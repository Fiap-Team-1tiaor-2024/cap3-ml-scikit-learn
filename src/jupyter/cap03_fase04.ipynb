# **Regressao Linear**

```
!pip install feature_engine
```

```
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from feature_engine.wrappers import SklearnTransformerWrapper
from sklearn.pipeline import Pipeline

######################################################
# Adicione o código da Regressão Linear via Pipeline #
######################################################
lr_model = Pipeline(steps=[('scaling', SklearnTransformerWrapper(transformer=StandardScaler(), variables=features)),
                           ('model', LinearRegression())])
```

```
# criando modelo
lr_model.fit(X_train, y_train)
```

```
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error

def adjusted_r2(y_esperado, y_previsto, X_treino):
  r2 = r2_score(y_esperado, y_previsto)
  N = len(y_esperado)
  p = X_treino.shape[1]

  r2_ajustado = (1 - ((1 - r2) * (N - 1)) / (N - p - 1))
  return r2_ajustado

def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred)/y_true))

def log_results(modelo_objeto, nome_modelo, X_train, X_test, y_train, y_test):
    y_pred_train = modelo_objeto.predict(X_train)
    y_pred_test  = modelo_objeto.predict(X_test)

    df_results = {'modelo': [], 'modo': [], 'r2': [], 'r2_ajustado': [], 'mse': [], 'rmse': [], 'rmsle': [], 'mae': [],
                  'medae': [], 'mape': []}

    # logando os resultados de treino
    df_results['modelo'].append(nome_modelo)
    df_results['modo'].append('treino')
    df_results['r2'].append(r2_score(y_train, y_pred_train))
    df_results['r2_ajustado'].append(adjusted_r2(y_train, y_pred_train, X_train))
    df_results['mse'].append(mean_squared_error(y_train, y_pred_train))
    df_results['rmse'].append(np.sqrt(mean_squared_error(y_train, y_pred_train)))
    df_results['rmsle'].append(np.sqrt(mean_squared_log_error(y_train, y_pred_train)))
    df_results['mae'].append(mean_absolute_error(y_train, y_pred_train))
    df_results['medae'].append(median_absolute_error(y_train, y_pred_train))
    df_results['mape'].append(mape(y_train, y_pred_train))

    # logando os resultados de teste
    df_results['modelo'].append(nome_modelo)
    df_results['modo'].append('teste')
    df_results['r2'].append(r2_score(y_test, y_pred_test))
    df_results['r2_ajustado'].append(adjusted_r2(y_test, y_pred_test, X_test))
    df_results['mse'].append(mean_squared_error(y_test, y_pred_test))
    df_results['rmse'].append(np.sqrt(mean_squared_error(y_test, y_pred_test)))
    df_results['rmsle'].append(np.sqrt(mean_squared_log_error(y_test, y_pred_test)))
    df_results['mae'].append(mean_absolute_error(y_test, y_pred_test))
    df_results['medae'].append(median_absolute_error(y_test, y_pred_test))
    df_results['mape'].append(mape(y_test, y_pred_test))

    return pd.DataFrame(df_results)
```

```
lr_results = log_results(lr_model, 'Regressão Linear', X_train, X_test, y_train, y_test)
```

```
lr_results = log_results(lr_model, 'Regressão Linear', X_train, X_test, y_train, y_test)
lr_results
```

# **Ridge Regression**

```
from sklearn.linear_model import Ridge

ridge_model = Pipeline(steps=[('scaling', SklearnTransformerWrapper(transformer=StandardScaler(), variables=features)),
                              ('model', Ridge())])

ridge_model.fit(X_train, y_train)

ridge_results = log_results(ridge_model, 'Ridge', X_train, X_test, y_train, y_test)
ridge_results
```

# **Decision Tree**

```
from sklearn.tree import DecisionTreeRegressor

tree_model = DecisionTreeRegressor(random_state=30, max_depth=3)

tree_model.fit(X_train, y_train)

tree_results = log_results(tree_model, 'Decision Tree', X_train, X_test, y_train, y_test)
tree_results
```

# **LGBM**

```
from lightgbm import LGBMRegressor

lgbm_model = LGBMRegressor(random_state=30)

lgbm_model.fit(X_train, y_train)

lgbm_results = log_results(lgbm_model, 'LightGBM', X_train, X_test, y_train, y_test)
lgbm_results
```
